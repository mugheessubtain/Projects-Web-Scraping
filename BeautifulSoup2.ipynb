```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://realpython.github.io/fake-jobs/"
req=requests.get(url)
req
```

```python
soup=BeautifulSoup(req.content,"html.parser")
soup
```

```python
title=soup.find_all("h2",class_="title is-5")
for tit in title:
    print(tit.text)
sub_title=soup.find("h3",class_="subtitle is-6 company")
location=soup.find("p",class_="location")
date=soup.find("p",class_="is-small has-text-grey")
# print(sub_title.text)
# print(location.text)
# print(date.text)

```

```python
allData=[]
# with open("allData.csv","w") as file:
  # file.write("Title,Sub-Title,Location,Date\n")
divs=soup.find_all("div",class_="column is-half")
for div in divs:
  # data=[]
  title=div.find("h2",class_="title is-5").text
  sub_title=div.find("h3",class_="subtitle is-6 company").text
  location=div.find("p",class_="location").text.strip()
  # cleanedLocation = location.replace('\n', '').replace('\t', '').strip()
  date=div.find("p",class_="is-small has-text-grey").text.strip()
  # data=[title,sub_title,location,date]
  # allData.append(data)
  # print(title)
  # print(sub_title)
  # print(location)
  # print(date
  jobs={
        "title":title,
        "sub_title":sub_title,
        "location":location,
        "date":date
    }
  allData.append(jobs)
    # file.write(f"{title},{sub_title},{location},{date}\n")


```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
path= r'C:\Users\hp\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe'
options = Options()
options.add_argument("--headless")  # Run in headless mode (no GUI)
driver= webdriver.Chrome(path,options=options)
# driver= webdriver.Chrome()

# driver.get("https://coinmarketcap.com/")
driver.get("https://realpython.github.io/fake-jobs/")
```

```python
title=driver.find_element(By.CLASS_NAME,"title.is-5").text
title
```

```python
title=driver.find_element(By.XPATH,r"//*[@id='ResultsContainer']/div[1]/div/div/div[1]/div[2]/h2").text
title
```

```python
divs=driver.find_elements(By.CLASS_NAME,"column.is-half")
for div in divs:
    title=div.find_element(By.CLASS_NAME,"title.is-5").text
    sub_title=div.find_element(By.CLASS_NAME,"subtitle.is-6.company").text
    location=div.find_element(By.CLASS_NAME,"location").text.strip()
    date=div.find_element(By.CLASS_NAME,"is-small.has-text-grey").text.strip()
    print(f"{title},{sub_title},'{location}','{date}'")
```
